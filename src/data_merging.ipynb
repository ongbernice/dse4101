```python
import pandas as pd
```

```python
# Merge final_data_for_modelling with stock_characteristics data
final_data = pd.read_csv("../data/final_data_for_modelling.csv")
final_data
```

```python
# Keep only 'Date', 'Ticker', 'Risk_Premium' (y variable) and columns containing 'lag'
columns_to_keep = ['Date', 'Ticker', 'Risk_Premium'] + [col for col in final_data.columns if 'lag' in col]
final_data = final_data[columns_to_keep]
final_data
```

```python
# Remove rows where the date is after December 2021 to align with characteristics data
final_data = final_data[final_data['Date'] <= '2021-12-01']
final_data
```

```python
stock_char = pd.read_csv("../data/characteristics_data.csv")
stock_char = stock_char[1:] # remove index column
stock_char = stock_char[stock_char['Date'] >= '1999-05-01'] # to align with final_data
stock_char
```

```python
# check whether number of tickers are the same
len(final_data['Ticker'].unique())
len(stock_char['Ticker'].unique())
```

```python
# filter final_data for tickers that are present in both datasets
stock_list = stock_char['Ticker'].unique().tolist()
final_data = final_data[final_data['Ticker'].isin(stock_list)]
final_data
```

```python
# drop duplicated entry in stock_char
print(final_data.duplicated(subset=['Date', 'Ticker']).sum())
print(stock_char.duplicated(subset=['Date', 'Ticker']).sum())
stock_char = stock_char.drop_duplicates(subset=['Date', 'Ticker'])
```

```python
# merge the two datasets
final_merged_data = final_data.merge(stock_char, on = ['Date', 'Ticker'], how = 'left')
final_merged_data = final_merged_data.drop(columns=['PERMNO'])
final_merged_data
```

```python
final_merged_data.to_csv("../data/final_merged_data.csv")
```
